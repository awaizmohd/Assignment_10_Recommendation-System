```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
book=pd.read_csv('book.csv',encoding='latin-1')
book
```

```python
book1 = book.drop("Unnamed: 0", axis=1)
book2 = book1.rename({"User.ID":"UserID", "Book.Title":"Title","Book.Rating":"Rating"}, axis = 1)
book2.head()
```

```python
book2.info()
```

```python
#number of unique users in the dataset
len(book2.UserID.unique())
```

```python
len(book2["Title"].unique())
```

```python
user_book = book2.pivot_table(index="UserID", columns="Title", values="Rating").reset_index(drop=True)
user_book
```

```python
user = book2["UserID"].unique()
user_df = pd.DataFrame(user, columns=["UserID"])
user_df
```

```python
user_book.index = user_df["UserID"].unique()
user_book
```

```python
#Impute those NaNs with 0 values
user_book.fillna(0, inplace=True)
user_book
```

# Cosine - based similarity

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim  = 1 - pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim 
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = user_df["UserID"].unique()
user_sim_df.columns= user_df["UserID"].unique()
user_sim_df

```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5,0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
